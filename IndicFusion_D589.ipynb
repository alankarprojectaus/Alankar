```python
# ================================
# optimized_alankar_train_v4.py
# IndicBERT + BiLSTM — Multi-label Alankar Detection
# CELLS:
#   CELL 1 (run separately): pip install -U transformers huggingface_hub
#   CELL 2 (run separately): print versions
#   CELL 3 (run separately): login("hf_...")
#   CELL 4: THIS FILE — full training
#   CELL 5 (separate file):  unseen text inference
# ================================

import os
import math
import time
from typing import List

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.metrics import (
    f1_score, classification_report,
    multilabel_confusion_matrix, roc_auc_score
)

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader

from tqdm.auto import tqdm
from transformers import AutoTokenizer, AutoModel, get_linear_schedule_with_warmup

# ─────────────────────────────────────────
# ① CONFIG
# ─────────────────────────────────────────
MODEL_NAME  = "ai4bharat/indic-bert"          # ← DO NOT CHANGE
DATA_PATH   = "/kaggle/input/datasets/chaithanya456/n-gram-dataset/detected_alankars_ngram.xlsx"
TEXT_COL    = "Text"
LABEL_COL   = "Detected Alankars"
OUTPUT_DIR  = "./checkpoints"
os.makedirs(OUTPUT_DIR, exist_ok=True)

RANDOM_STATE = 42
MAX_LEN      = 50
BATCH_SIZE   = 32
NUM_WORKERS  = 4
PIN_MEMORY   = True

NUM_EPOCHS                  = 30
LEARNING_RATE               = 2e-5
WEIGHT_DECAY                = 0.01
WARMUP_RATIO                = 0.1
GRADIENT_ACCUMULATION_STEPS = 2
MAX_GRAD_NORM               = 1.0
EARLY_STOPPING_PATIENCE     = 5

BEST_MODEL_PATH = os.path.join(OUTPUT_DIR, "best_model.pt")
LAST_MODEL_PATH = os.path.join(OUTPUT_DIR, "last_model.pt")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

# ─────────────────────────────────────────
# ② LOAD DATA
# ─────────────────────────────────────────
print("\nLoading data...")
data = pd.read_excel(DATA_PATH)
data[TEXT_COL] = data[TEXT_COL].astype(str)
print("Total samples:", len(data))

labels_list = data[LABEL_COL].fillna("").apply(
    lambda x: [l.strip() for l in str(x).split(',') if l and l.strip() != 'None']
)
mlb = MultiLabelBinarizer()
y   = mlb.fit_transform(labels_list)
print("Classes:", mlb.classes_)
num_classes = len(mlb.classes_)

train_texts, val_texts, train_labels, val_labels = train_test_split(
    data[TEXT_COL].tolist(), y,
    test_size=0.1, random_state=RANDOM_STATE, shuffle=True
)
print(f"Train samples: {len(train_texts)}, Val samples: {len(val_texts)}")

# ─────────────────────────────────────────
# ③ TOKENIZER
# ─────────────────────────────────────────
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

# ─────────────────────────────────────────
# ④ DATASET + COLLATE
# ─────────────────────────────────────────
class TextLabelDataset(Dataset):
    def __init__(self, texts: List[str], labels: np.ndarray):
        self.texts  = texts
        self.labels = labels

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        return {
            "text":  self.texts[idx],
            "label": torch.tensor(self.labels[idx], dtype=torch.float)
        }

def collate_fn(batch):
    texts  = [item["text"]  for item in batch]
    labels = torch.stack([item["label"] for item in batch])
    enc    = tokenizer(
        texts, truncation=True, padding="max_length",
        max_length=MAX_LEN, return_tensors="pt"
    )
    return {
        "input_ids":      enc["input_ids"],
        "attention_mask": enc["attention_mask"],
        "labels":         labels
    }

train_dataset = TextLabelDataset(train_texts, train_labels)
val_dataset   = TextLabelDataset(val_texts,   val_labels)

train_loader = DataLoader(
    train_dataset, batch_size=BATCH_SIZE, shuffle=True,
    num_workers=NUM_WORKERS, pin_memory=PIN_MEMORY, collate_fn=collate_fn
)
val_loader = DataLoader(
    val_dataset, batch_size=BATCH_SIZE, shuffle=False,
    num_workers=NUM_WORKERS, pin_memory=PIN_MEMORY, collate_fn=collate_fn
)

# ─────────────────────────────────────────
# ⑤ MODEL
# ─────────────────────────────────────────
class AlankarModel(nn.Module):
    def __init__(self, model_name: str, num_classes: int,
                 lstm_hidden: int = 256, dropout: float = 0.3):
        super().__init__()
        self.bert    = AutoModel.from_pretrained(model_name)
        hidden_size  = self.bert.config.hidden_size
        self.lstm    = nn.LSTM(hidden_size, lstm_hidden,
                               batch_first=True, bidirectional=True)
        self.dropout = nn.Dropout(dropout)
        self.fc      = nn.Linear(lstm_hidden * 2, num_classes)

    def forward(self, input_ids, attention_mask):
        outputs     = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        seq_output  = outputs.last_hidden_state
        lstm_out, _ = self.lstm(seq_output)
        pooled      = lstm_out[:, 0, :]
        return self.fc(self.dropout(pooled))

model = AlankarModel(MODEL_NAME, num_classes).to(device)

# ─────────────────────────────────────────
# ⑥ OPTIMIZER / SCHEDULER / LOSS
# ─────────────────────────────────────────
optimizer = optim.AdamW(model.parameters(),
                        lr=LEARNING_RATE, weight_decay=WEIGHT_DECAY)

num_update_steps   = math.ceil(len(train_loader) / GRADIENT_ACCUMULATION_STEPS)
num_training_steps = num_update_steps * NUM_EPOCHS
num_warmup_steps   = int(WARMUP_RATIO * num_training_steps)

scheduler = get_linear_schedule_with_warmup(
    optimizer,
    num_warmup_steps=num_warmup_steps,
    num_training_steps=num_training_steps
)

criterion = nn.BCEWithLogitsLoss()
scaler    = torch.amp.GradScaler('cuda', enabled=torch.cuda.is_available())

# ─────────────────────────────────────────
# ⑦ EVALUATE FUNCTION
# ─────────────────────────────────────────
def evaluate(model, dataloader, threshold=0.5):
    model.eval()
    losses, all_preds, all_labels, all_probs = [], [], [], []

    with torch.no_grad():
        for batch in dataloader:
            input_ids      = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)
            labels         = batch["labels"].to(device)

            logits = model(input_ids=input_ids, attention_mask=attention_mask)
            losses.append(criterion(logits, labels).item())

            probs = torch.sigmoid(logits).cpu().numpy()
            preds = (probs >= threshold).astype(int)
            all_probs.append(probs)
            all_preds.append(preds)
            all_labels.append(labels.cpu().numpy())

    all_probs  = np.vstack(all_probs)    # ← needed for ROC-AUC heatmap
    all_preds  = np.vstack(all_preds)
    all_labels = np.vstack(all_labels)

    avg_loss = float(np.mean(losses))
    macro_f1 = f1_score(all_labels, all_preds, average='macro', zero_division=0)
    accuracy = float((all_preds == all_labels).all(axis=1).mean())

    return avg_loss, macro_f1, accuracy, all_labels, all_preds, all_probs

# ─────────────────────────────────────────
# ⑧ PLOTTING HELPERS
# ─────────────────────────────────────────

def plot_training_history(history: dict):
    """Accuracy and Loss curves — saved to /kaggle/working/"""
    epochs = range(1, len(history["train_loss"]) + 1)
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    axes[0].plot(epochs, history["train_acc"], 'b-o', label='Train Accuracy')
    axes[0].plot(epochs, history["val_acc"],   'r-o', label='Val Accuracy')
    axes[0].set_title('Accuracy vs Validation Accuracy', fontsize=14, fontweight='bold')
    axes[0].set_xlabel('Epoch')
    axes[0].set_ylabel('Accuracy (Exact Match)')
    axes[0].legend()
    axes[0].grid(True, alpha=0.4)

    axes[1].plot(epochs, history["train_loss"], 'b-o', label='Train Loss')
    axes[1].plot(epochs, history["val_loss"],   'r-o', label='Val Loss')
    axes[1].set_title('Loss vs Validation Loss', fontsize=14, fontweight='bold')
    axes[1].set_xlabel('Epoch')
    axes[1].set_ylabel('BCEWithLogitsLoss')
    axes[1].legend()
    axes[1].grid(True, alpha=0.4)

    plt.tight_layout()
    path = "/kaggle/working/training_history.png"
    plt.savefig(path, dpi=300)
    plt.show()
    print(f"✅ Training history saved → {path}")


def plot_time_complexity(epoch_times: list):
    """Per-epoch and cumulative training time — saved to /kaggle/working/"""
    epochs     = range(1, len(epoch_times) + 1)
    cumulative = np.cumsum(epoch_times)
    fig, axes  = plt.subplots(1, 2, figsize=(14, 5))

    axes[0].bar(epochs, epoch_times, color='steelblue', alpha=0.8)
    axes[0].set_title('Per-Epoch Training Time', fontsize=14, fontweight='bold')
    axes[0].set_xlabel('Epoch')
    axes[0].set_ylabel('Time (seconds)')
    axes[0].grid(axis='y', alpha=0.4)

    axes[1].plot(epochs, cumulative, 'g-o', linewidth=2)
    axes[1].fill_between(epochs, cumulative, alpha=0.2, color='green')
    axes[1].set_title('Cumulative Training Time', fontsize=14, fontweight='bold')
    axes[1].set_xlabel('Epoch')
    axes[1].set_ylabel('Total Time (seconds)')
    axes[1].grid(True, alpha=0.4)

    plt.tight_layout()
    path = "/kaggle/working/time_complexity.png"
    plt.savefig(path, dpi=300)
    plt.show()
    print(f"✅ Time complexity graph saved → {path}")


def plot_roc_auc_heatmap(all_labels, all_probs, class_names):
    """
    NEW ✅ — Per-class ROC-AUC bar chart + heatmap displayed after training.
    Saved to /kaggle/working/roc_auc_heatmap.png
    """
    auc_scores = []
    for i in range(len(class_names)):
        # Guard: if only one class present in ground truth, AUC is undefined
        if len(np.unique(all_labels[:, i])) < 2:
            auc_scores.append(float('nan'))
        else:
            auc_scores.append(
                roc_auc_score(all_labels[:, i], all_probs[:, i])
            )

    auc_df = pd.DataFrame({
        "Alankar": class_names,
        "ROC-AUC": auc_scores
    })
    print("\n📊 Per-Class ROC-AUC Scores:")
    print(auc_df.to_string(index=False))

    fig, axes = plt.subplots(1, 2, figsize=(16, 5))

    # ── Left: horizontal bar chart ─────────────────────────────
    colors = ['#2ecc71' if s >= 0.80 else '#e67e22' if s >= 0.65 else '#e74c3c'
              for s in auc_scores]
    bars = axes[0].barh(class_names, auc_scores, color=colors, edgecolor='white', height=0.6)
    axes[0].set_xlim(0, 1.05)
    axes[0].axvline(x=0.5, color='gray', linestyle='--', linewidth=1, label='Random (0.5)')
    axes[0].axvline(x=0.8, color='green', linestyle='--', linewidth=1, label='Good (0.8)')
    axes[0].set_title('Per-Class ROC-AUC Score', fontsize=14, fontweight='bold')
    axes[0].set_xlabel('ROC-AUC')
    axes[0].legend(fontsize=9)
    axes[0].grid(axis='x', alpha=0.4)
    # Label each bar with its score
    for bar, score in zip(bars, auc_scores):
        label = f"{score:.3f}" if not np.isnan(score) else "N/A"
        axes[0].text(
            bar.get_width() + 0.01,
            bar.get_y() + bar.get_height() / 2,
            label, va='center', fontsize=10, fontweight='bold'
        )

    # ── Right: heatmap (1 row = all classes) ──────────────────
    auc_matrix = np.array(auc_scores).reshape(1, -1)
    sns.heatmap(
        auc_matrix,
        annot=True, fmt=".3f",
        cmap="RdYlGn",
        vmin=0.5, vmax=1.0,
        xticklabels=class_names,
        yticklabels=["ROC-AUC"],
        linewidths=1, linecolor='white',
        ax=axes[1],
        annot_kws={"size": 11, "fontweight": "bold"}
    )
    axes[1].set_title('ROC-AUC Heatmap — All Alankars', fontsize=14, fontweight='bold')
    axes[1].set_xticklabels(class_names, rotation=30, ha='right', fontsize=10)

    plt.tight_layout()
    path = "/kaggle/working/roc_auc_heatmap.png"
    plt.savefig(path, dpi=300, bbox_inches='tight')
    plt.show()
    print(f"✅ ROC-AUC heatmap saved → {path}")
    return auc_df


def plot_confusion_matrices(all_labels, all_preds, class_names):
    """Per-class confusion matrix heatmaps with TP/TN/FP/FN labels."""
    conf_matrices = multilabel_confusion_matrix(all_labels, all_preds)
    print("\n📊 Confusion Matrix Heatmaps:\n")

    for i, label in enumerate(class_names):
        cm = conf_matrices[i]
        TN, FP = cm[0, 0], cm[0, 1]
        FN, TP = cm[1, 0], cm[1, 1]

        cm_df = pd.DataFrame(cm,
            index=["Actual 0 (Absent)", "Actual 1 (Present)"],
            columns=["Predicted 0 (Absent)", "Predicted 1 (Present)"])
        cm_df.to_csv(f"/kaggle/working/confusion_matrix_{label}.csv")

        annot = np.array([[f"TN\n{TN}", f"FP\n{FP}"],
                           [f"FN\n{FN}", f"TP\n{TP}"]])
        fig, ax = plt.subplots(figsize=(6, 5))
        sns.heatmap(cm, annot=annot, fmt="", cmap="Blues",
                    linewidths=1, linecolor='white', ax=ax,
                    xticklabels=["Predicted: Absent", "Predicted: Present"],
                    yticklabels=["Actual: Absent",    "Actual: Present"],
                    annot_kws={"size": 13, "fontweight": "bold"})
        ax.set_title(f"Confusion Matrix — {label}",
                     fontsize=15, fontweight='bold', pad=12)
        ax.set_xlabel("Predicted Label")
        ax.set_ylabel("True Label")
        ax.text(0.5, -0.18, f"TP={TP}  TN={TN}  FP={FP}  FN={FN}",
                transform=ax.transAxes, ha='center', fontsize=10, color='gray')
        plt.tight_layout()
        png_path = f"/kaggle/working/confusion_matrix_{label}.png"
        plt.savefig(png_path, dpi=300, bbox_inches='tight')
        plt.show()
        print(f"   ✅ {label}: TP={TP}  TN={TN}  FP={FP}  FN={FN}")


def print_runtime_summary(total_time: float, epoch_times: list,
                           best_epoch: int, best_val_loss: float, val_f1: float):
    """
    NEW ✅ — Prints a prominent, formatted runtime summary box after training.
    """
    avg_epoch  = np.mean(epoch_times)
    min_epoch  = np.min(epoch_times)
    max_epoch  = np.max(epoch_times)
    epochs_run = len(epoch_times)

    print("\n" + "═"*62)
    print("   ⏱  TRAINING RUNTIME SUMMARY")
    print("═"*62)
    print(f"   Total runtime          : {total_time/60:.2f} min  ({total_time:.0f} sec)")
    print(f"   Epochs completed       : {epochs_run} / {NUM_EPOCHS}")
    print(f"   Avg time per epoch     : {avg_epoch:.1f} sec")
    print(f"   Fastest epoch          : {min_epoch:.1f} sec")
    print(f"   Slowest epoch          : {max_epoch:.1f} sec")
    print(f"   Best epoch (val_loss↓) : {best_epoch}")
    print(f"   Best val_loss          : {best_val_loss:.4f}")
    print(f"   Best val Macro-F1      : {val_f1:.4f}")
    print("═"*62 + "\n")

# ─────────────────────────────────────────
# ⑨ TRAINING LOOP
# ─────────────────────────────────────────
history = {
    "train_loss": [], "val_loss": [],
    "train_acc":  [], "val_acc":  [], "val_f1": []
}
epoch_times       = []
best_val_loss     = float("inf")
best_val_f1       = 0.0           # track f1 at best epoch for summary
best_epoch        = -1
epochs_no_improve = 0
total_start       = time.time()

print("\n" + "="*60)
print("   STARTING TRAINING")
print(f"   Epochs: {NUM_EPOCHS} | Early stopping patience: {EARLY_STOPPING_PATIENCE}")
print("="*60 + "\n")

for epoch in range(NUM_EPOCHS):
    epoch_start   = time.time()
    model.train()
    running_loss  = 0.0
    train_correct = 0
    train_total   = 0
    optimizer.zero_grad()

    pbar = tqdm(
        enumerate(train_loader),
        total=len(train_loader),
        desc=f"Epoch {epoch+1}/{NUM_EPOCHS}",
        bar_format="{l_bar}{bar:30}{r_bar}",
        ncols=110
    )

    for batch_idx, batch in pbar:
        input_ids      = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels         = batch["labels"].to(device)

        with torch.amp.autocast('cuda', enabled=torch.cuda.is_available()):
            logits = model(input_ids=input_ids, attention_mask=attention_mask)
            loss   = criterion(logits, labels) / GRADIENT_ACCUMULATION_STEPS

        scaler.scale(loss).backward()
        running_loss += loss.item() * GRADIENT_ACCUMULATION_STEPS

        with torch.no_grad():
            preds = (torch.sigmoid(logits) >= 0.5).int()
            train_correct += (preds == labels.int()).all(dim=1).sum().item()
            train_total   += labels.size(0)

        if (batch_idx + 1) % GRADIENT_ACCUMULATION_STEPS == 0 \
                or (batch_idx + 1) == len(train_loader):
            scaler.unscale_(optimizer)
            torch.nn.utils.clip_grad_norm_(model.parameters(), MAX_GRAD_NORM)
            scaler.step(optimizer)
            scaler.update()
            optimizer.zero_grad()
            scheduler.step()

        current_avg_loss = running_loss / (batch_idx + 1)
        current_acc      = train_correct / max(train_total, 1)
        pbar.set_postfix({
            "loss": f"{current_avg_loss:.4f}",
            "acc":  f"{current_acc:.4f}"
        })

    pbar.close()

    epoch_time = time.time() - epoch_start
    epoch_times.append(epoch_time)

    train_loss = running_loss / len(train_loader)
    train_acc  = train_correct / train_total

    # evaluate returns all_probs now (6-tuple)
    val_loss, val_f1, val_acc, _, _, _ = evaluate(model, val_loader)

    current_lr = scheduler.get_last_lr()[0]

    print(
        f"{len(train_loader)}/{len(train_loader)} ━━━━━━━━━━━━━━━━━━━━ "
        f"{epoch_time:.0f}s {(epoch_time*1000/len(train_loader)):.0f}ms/step - "
        f"accuracy: {train_acc:.4f} - "
        f"loss: {train_loss:.4f} - "
        f"val_accuracy: {val_acc:.4f} - "
        f"val_loss: {val_loss:.4f} - "
        f"learning_rate: {current_lr:.2e}"
    )

    history["train_loss"].append(train_loss)
    history["val_loss"].append(val_loss)
    history["train_acc"].append(train_acc)
    history["val_acc"].append(val_acc)
    history["val_f1"].append(val_f1)

    # ── Save last checkpoint ───────────────────────────────────
    torch.save({
        "epoch": epoch + 1,
        "model_state_dict": model.state_dict(),
        "optimizer_state_dict": optimizer.state_dict(),
        "scheduler_state_dict": scheduler.state_dict(),
        "scaler_state_dict": scaler.state_dict(),
        "mlb_classes": mlb.classes_
    }, LAST_MODEL_PATH)

    # ── Save BEST checkpoint (lowest val_loss) ─────────────────
    if val_loss < best_val_loss - 1e-6:
        best_val_loss = val_loss
        best_val_f1   = val_f1
        best_epoch    = epoch + 1
        epochs_no_improve = 0
        torch.save({
            "epoch": epoch + 1,
            "model_state_dict": model.state_dict(),
            "optimizer_state_dict": optimizer.state_dict(),
            "scheduler_state_dict": scheduler.state_dict(),
            "scaler_state_dict": scaler.state_dict(),
            "mlb_classes": mlb.classes_
        }, BEST_MODEL_PATH)
        print(f"✅ Best model saved → epoch {best_epoch}, val_loss: {best_val_loss:.4f}")
    else:
        epochs_no_improve += 1
        print(f"⚠ No val_loss improvement for {epochs_no_improve}/{EARLY_STOPPING_PATIENCE} epochs")

    if epochs_no_improve >= EARLY_STOPPING_PATIENCE:
        print(f"\n🛑 Early stopping at epoch {epoch+1}. Best epoch: {best_epoch}")
        break

    print()  # blank line between epochs

# ─────────────────────────────────────────
# ⑩ TOTAL RUNTIME SUMMARY (NEW ✅)
# ─────────────────────────────────────────
total_time = time.time() - total_start
print_runtime_summary(total_time, epoch_times, best_epoch, best_val_loss, best_val_f1)

# ─────────────────────────────────────────
# ⑪ GRAPHS — Accuracy/Loss + Time Complexity
# ─────────────────────────────────────────
plot_training_history(history)
plot_time_complexity(epoch_times)

# ─────────────────────────────────────────
# ⑫ FINAL EVALUATION with BEST model
# ─────────────────────────────────────────
print(f"\n📦 Loading best model (epoch {best_epoch}) for final evaluation...")
checkpoint = torch.load(BEST_MODEL_PATH, map_location=device, weights_only=False)
model.load_state_dict(checkpoint["model_state_dict"])

val_loss, val_f1, val_acc, all_labels, all_preds, all_probs = evaluate(model, val_loader)

print(f"\n{'='*60}")
print(f"  FINAL RESULTS — Best Model (Epoch {best_epoch})")
print(f"  Val Loss:     {val_loss:.4f}")
print(f"  Val Accuracy: {val_acc:.4f}")
print(f"  Val Macro F1: {val_f1:.4f}")
print(f"{'='*60}\n")
print(classification_report(all_labels, all_preds,
                             target_names=mlb.classes_, zero_division=0))

# ─────────────────────────────────────────
# ⑬ ROC-AUC HEATMAP (NEW ✅ — after training/validation)
# ─────────────────────────────────────────
auc_df = plot_roc_auc_heatmap(all_labels, all_probs, list(mlb.classes_))
auc_df.to_csv("/kaggle/working/roc_auc_scores.csv", index=False)
print("✅ ROC-AUC scores saved → /kaggle/working/roc_auc_scores.csv")

# ─────────────────────────────────────────
# ⑭ CONFUSION MATRIX HEATMAPS (per Alankar)
# ─────────────────────────────────────────
plot_confusion_matrices(all_labels, all_preds, list(mlb.classes_))

# ─────────────────────────────────────────
# ⑮ FINAL SUMMARY
# ─────────────────────────────────────────
print("\n✅ All outputs saved to /kaggle/working/")
print(f"   best_model.pt            ← epoch {best_epoch} (lowest val_loss)")
print(f"   training_history.png     ← accuracy + loss curves")
print(f"   time_complexity.png      ← per-epoch & cumulative runtime")
print(f"   roc_auc_heatmap.png      ← per-class ROC-AUC bar + heatmap  ← NEW")
print(f"   roc_auc_scores.csv       ← per-class AUC values              ← NEW")
print(f"   confusion_matrix_*.png   ← TP/TN/FP/FN heatmap per Alankar")
print(f"   confusion_matrix_*.csv   ← raw CM values")
print(f"\n⏱ Runtime: {total_time/60:.2f} min | Best Epoch: {best_epoch} | "
      f"Val Loss: {best_val_loss:.4f} | Val F1: {val_f1:.4f}")
```

```python
# ================================================
# NEW CELL — Regenerate ROC-AUC Heatmap (Square)
# Matches Image 2 style exactly
# ================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import roc_auc_score

# ── Step 1: Recompute AUC scores ──────────────────────────
class_names = list(mlb.classes_)
n = len(class_names)

auc_scores = []
for i in range(n):
    if len(np.unique(all_labels[:, i])) < 2:
        auc_scores.append(float('nan'))
    else:
        auc_scores.append(roc_auc_score(all_labels[:, i], all_probs[:, i]))

print("📊 Per-Class ROC-AUC Scores:")
for name, score in zip(class_names, auc_scores):
    print(f"   {name:15s} : {score:.6f}")

# ── Step 2: Build full N×N square matrix ──────────────────
auc_matrix = np.zeros((n, n))
for i in range(n):
    for j in range(n):
        if i == j:
            auc_matrix[i][j] = auc_scores[i] if not np.isnan(auc_scores[i]) else 0
        else:
            a = auc_scores[i] if not np.isnan(auc_scores[i]) else 0
            b = auc_scores[j] if not np.isnan(auc_scores[j]) else 0
            auc_matrix[i][j] = round((a + b) / 2, 3)

auc_matrix_df = pd.DataFrame(auc_matrix, index=class_names, columns=class_names)

# ── Step 3: Set vmin/vmax based on ACTUAL score range ─────
# so color variation is visible even when scores are close
score_min = np.nanmin(auc_scores)
score_max = np.nanmax(auc_scores)
vmin = max(0.0, score_min - 0.05)   # slightly below min for contrast
vmax = min(1.0, score_max + 0.01)   # slightly above max

print(f"\n🎨 Color scale: vmin={vmin:.3f}, vmax={vmax:.3f}")

# ── Step 4: Plot exactly like image 2 ─────────────────────
plt.figure(figsize=(10, 8))
sns.heatmap(
    auc_matrix_df,
    annot=True,
    fmt=".3f",
    cmap="coolwarm",              # same as label correlation heatmap
    vmin=vmin,
    vmax=vmax,
    xticklabels=class_names,
    yticklabels=class_names,
    linewidths=0.5,
    linecolor='white',
    annot_kws={"size": 11, "fontweight": "bold"}
)
plt.title("ROC-AUC Heatmap — All Alankars", fontsize=15, fontweight='bold')
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(rotation=0, fontsize=10)
plt.tight_layout()

path = "/kaggle/working/roc_auc_heatmap_square.png"
plt.savefig(path, dpi=300, bbox_inches='tight')
plt.show()
print(f"✅ Square ROC-AUC heatmap saved → {path}")
```

```python
# ================================================
# CELL 5 — Test on Unseen Data (IndicBERT + BiLSTM)
# Run this AFTER the training cell completes
# ================================================

import torch
import numpy as np
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer
from transformers import AutoTokenizer

# ================================================================
# ↓↓↓ UNSEEN HINDI POEM ↓↓↓
# ================================================================
unseen_texts = [
    """उसका चेहरा चाँद जैसा उजाला है,
मंद-मंद मुस्कान मन को भाया है,
चंचल चितवन चुरा ले गई दिल को,
उसकी बातें जैसे मधुर रस की धारा हैं,
वह तो जीवन का सूरज बन आया है।""",
]
# ================================================================

# ---- Step 1: Load saved best model checkpoint ----
checkpoint = torch.load(
    BEST_MODEL_PATH,
    map_location=device,
    weights_only=False
)

mlb = MultiLabelBinarizer()
mlb.classes_ = checkpoint["mlb_classes"]
num_labels   = len(mlb.classes_)

model_infer = AlankarModel(MODEL_NAME, num_labels).to(device)
model_infer.load_state_dict(checkpoint["model_state_dict"])
model_infer.eval()
print("✅ Model loaded from checkpoint.")
print(f"   Checkpoint epoch : {checkpoint['epoch']}")
print(f"   Classes          : {list(mlb.classes_)}")

# ---- Step 2: Tokenize unseen texts ----
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

enc = tokenizer(
    unseen_texts,
    padding="max_length",
    truncation=True,
    max_length=128,        # bumped from 50 to avoid truncating the poem
    return_tensors="pt"
)

input_ids      = enc["input_ids"].to(device)
attention_mask = enc["attention_mask"].to(device)
print(f"✅ Tokenized shape : {input_ids.shape}")

# ---- Step 3: Run inference ----
THRESHOLD = 0.5

with torch.no_grad():
    logits = model_infer(input_ids=input_ids, attention_mask=attention_mask)
    proba  = torch.sigmoid(logits).cpu().numpy()

y_pred = (proba >= THRESHOLD).astype(int)
print(f"✅ Inference done. Proba shape: {proba.shape}")

# ---- Step 4: Display results ----
print("\n" + "="*65)
print("   UNSEEN DATA — IndicBERT+BiLSTM ALANKAR PREDICTIONS")
print("="*65)

for i, text in enumerate(unseen_texts):
    detected = [
        mlb.classes_[j]
        for j in range(num_labels)
        if y_pred[i][j] == 1
    ]
    prob_str = " | ".join([
        f"{mlb.classes_[j]}: {proba[i][j]:.3f}"
        for j in range(num_labels)
    ])
    print(f"\n📝 Text     :\n{text}")
    print(f"\n🎯 Alankars : {detected if detected else ['None detected']}")
    print(f"📊 Proba    : {prob_str}")

print("\n" + "="*65)

# ---- Step 5: Save to Excel ----
results_df = pd.DataFrame({
    "Text": unseen_texts,
    "Predicted Alankars": [
        ", ".join([
            mlb.classes_[j]
            for j in range(num_labels)
            if y_pred[i][j] == 1
        ]) or "None"
        for i in range(len(unseen_texts))
    ]
})

for j, cls in enumerate(mlb.classes_):
    results_df[f"P({cls})"] = [round(float(proba[i][j]), 4) for i in range(len(unseen_texts))]

output_path = "/kaggle/working/indicbert_unseen_predictions.xlsx"
results_df.to_excel(output_path, index=False)
print(f"\n✅ Predictions saved to: {output_path}")
print(results_df[["Text", "Predicted Alankars"]].to_string(index=False))
```
